```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate linear data
x = np.linspace(0, 10, 100)
y = 2 * x + 1 + np.random.normal(0, 2, 100)

# Plot the data
plt.scatter(x, y, label='y = 2x + 1')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Data')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# Reshape x to be a 2D array
x_reshaped = x.reshape(-1, 1)

# Create and fit the model
model = LinearRegression()
model.fit(x_reshaped, y)

# Get the coefficients
W = model.coef_[0]
b = model.intercept_

print(f"W: {W}, b: {b}")

# Plot the data and the fitted line
plt.scatter(x, y, label='Data')
plt.plot(x, W * x + b, color='red', label='Fitted line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Fit')
plt.legend()
plt.show()
```

```python
from ipywidgets import interact, widgets

def plot_manual_fit_with_errors(W, b):
    # Calculate predicted values
    y_pred = W * x + b
    
    # Calculate residuals
    residuals = y - y_pred
    
    # Create the plot
    plt.figure(figsize=(12, 6))
    
    # Plot the data points and fitted line
    plt.scatter(x, y, label='Data', alpha=0.6)
    plt.plot(x, y_pred, color='red', label='Manual fit')
     
    # Plot the error lines
    for i in range(len(x)):
        plt.vlines(x[i], y[i], y_pred[i], colors='gray', alpha=0.3)
    
    # Add labels and title
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Manual Linear Fit with Errors')
    plt.legend()
    
    # Show mean squared error
    mse = np.mean(residuals**2)
    plt.text(0.02, 0.98, f'MSE: {mse:.2f}', 
             transform=plt.gca().transAxes, 
             verticalalignment='top')
    
    plt.show()

interact(plot_manual_fit_with_errors, 
        W=widgets.FloatSlider(min=0, max=4, step=0.01, value=model.coef_[0]), 
        b=widgets.FloatSlider(min=-5, max=5, step=0.01, value=model.intercept_))
```
